### Essential Libraries

Let us import the necessary Python libraries.

> Pandas : Library for Data Acquisition and Preparation  
> Matplotlib : Low-level library for Data Visualization  
> Sklearn : Library for Machine Learning

```python
# Necessary imports
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt

from sklearn.metrics import confusion_matrix, accuracy_score, f1_score, roc_curve, auc, roc_auc_score
from sklearn.tree import plot_tree, DecisionTreeClassifier
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
sb.set_theme()

randomStateNumber = 20
```

```python
label = "DepDel15"

def setup(path):
    # Read data from file
    df = pd.read_csv(path)
    # Extract features and label set
    feature_set = pd.DataFrame(df.drop(labels=[label], axis=1))
    label_set = pd.DataFrame(df[label])
    return df, feature_set, label_set
```

```python
df, feature_set, label_set = setup('data/processed_dataset.csv')
df.head()
```

---

## 1. Preparing Train/Test Splits <a class="anchor" id="first-bullet"></a>

Here, we split our data into train and test splits so that we can teach our model using the `training` set and thereafter test its performance on unseen data, namely our `test` set. This ensures that our model does not 'memorise' the training data instead of generalising from it.

Instead of taking the first 80% of entries, we employ `train_test_split` from `sklearn` to form our splits through random sampling to ensure fairness.

```python
# Split our data into train and test splits by 80-20 ratio, using random sampling.
X_train, X_test, y_train, y_test = train_test_split(feature_set, label_set, test_size=0.2, random_state=randomStateNumber)
```

---

## 2. Training ML Models <a class="anchor" id="second-bullet"></a>

For a binary classification task, there are many popular models available to us. For this project, we have selected the following models for various reasons:


### **Decision Tree Classifier** 

`DecisionTreeClassifier` works by recursively splitting the dataset into subsets based on the most significant feature at each node of the tree. However, one thing to note is that `DecisionTreeClassifier` is **prone to overfitting**, and with the large dataset that we have, it might lead to poor generalisation performance as the model will learn to memorise training data instead of finding underlying patterns.<br><br>We have selected it for its **simplicity and ease of interpretation** as well as **ability to handle both numerical and categorical data without requiring feature scaling**.

```python
# Decision Tree Classifier
model_dt = DecisionTreeClassifier(random_state=randomStateNumber)
model_dt.fit(X_train, y_train)
model_dt_y_train_pred = model_dt.predict(X_train)
model_dt_y_test_pred = model_dt.predict(X_test)
```

---

### **Random Forest Classifier**

`RandomForestClassifier` is an ensemble learning method that combines multiple decision trees to improve the performance of the model and reduce overfitting. It works by constructing many decision trees during training, each trained on a random subset of the training data and a random subset of features. The final prediction is then made by aggregating the predictions of all individual trees. <br><br>We have selected to use `RandomForestClassifier` for its **robustness and scability** and that it is **less prone to overfitting**.

```python
# Random Forest Classifier
model_rf = RandomForestClassifier(random_state=randomStateNumber)
model_rf.fit(X_train, y_train.values.ravel())
model_rf_y_train_pred = model_rf.predict(X_train)
model_rf_y_test_pred = model_rf.predict(X_test)
```

---

### **Logistic Regression**

`LogisticRegression` is a linear model that is also commonly used for binary classification. It models the probability that a given input belongs to a particular class using the sigmoid function. `LogisticRegression` is known for being **easily interpretable** and **efficient to compute**, making it suitable for our use case. However, it may **struggle to capture complex nonlinear relationships** in the data. <br><br> We have selected `LinearRegression` for its **simplicity** as well.

```python
# Logistic Regression
model_lr = LogisticRegression()
model_lr.fit(X_train, y_train.values.ravel())
model_lr_y_train_pred = model_lr.predict(X_train)
model_lr_y_test_pred = model_lr.predict(X_test)
```

## 3. Metrics Selection <a class="anchor" id="third-bullet"></a>

To evaluate the models that we have trained, we have selected the following metrics:

### **Accuracy**
Accuracy measures the **proportion of correctly classified instances out of the total instances**.  <br>

<div class="alert alert-block alert-info">
Note: Accuracy though intuitive, <b>can be misleading for imbalanced datasets</b> as it may not reflect the true performance of the model.</div>

---
### **Confusion Matrix**
The confusion matrix shows a summary of the model's predictions compared to the actual truth across different classes.

| Confusion Matrix  |       |        |        |      
| :---              | :---: | :----: | :----: |         
| Actual Negative   |  (0)  |   TN   |   FP   |             
| Actual Positive   |  (1)  |   FN   |   TP   |       
|                   |       |   (0)   |   (1)   |       
|                   |       | Predicted Negative    |   Predicted Postitive  |   

* `TPR = TP / (TP + FN)` : True Positive Rate = True Positives / All Positives    
* `TNR = TN / (TN + FP)` : True Negative Rate = True Negatives / All Negatives    

* `FPR = FP / (TN + FP)` : False Positive Rate = False Positives / All Negatives 
* `FNR = FN / (TP + FN)` : False Negative Rate = False Negatives / All Positives 

* `Precision` : The proportion of **true positive predictions** amongst **all positive predictions** (TP / TP+FP)
* `Recall (Sensitivity)` : The proportion of **actual positive instances that are correctly identified by the model** 
---
### **F1-Score**
The harmonic mean of precision and recall, 1 indicates best performance, 0 indicates worst (2 x Precision x Recall / Precision + Recall).
F1-Score is useful when there is an imbalance between classes or where both false positives and false negatives need to be minimized. <br>
<div class="alert alert-block alert-success">
In the case of flight delays, we want to <b>minimize False Negatives</b> where true delays are missed out on while <b>keeping False Positives at an acceptable rate</b> (false alarm delays).
</div>

---
### **ROC-AUC**

The ROC curve is a popular performance metric used to evaluate the performance of binary classification models. 
The curve plots the `True Positive Rate (TPR)` against the `False Positive Rate (FPR)` at various threshold settings.

The `Area Under Curve (AOC)` indicates discriminantion between positive and negative classes.
The **higher** the value, the **better separation that is between positive and negative classes**.

---

### Helper Function

To make things easier, we consolidate all the metrics that we are interested in into a reusable function.


```python
def printMetrics(model_scores, graph_label, label, actual, pred):

    accuracy = accuracy_score(actual, pred)
    cmTrain = confusion_matrix(actual, pred)
    tpTrain = cmTrain[1][1] # True Positives : Good (1) predicted Good (1)
    fpTrain = cmTrain[0][1] # False Positives : Bad (0) predicted Good (1)
    tnTrain = cmTrain[0][0] # True Negatives : Bad (0) predicted Bad (0)
    fnTrain = cmTrain[1][0] # False Negatives : Good (1) predicted Bad (0)
    f1 = f1_score(actual, pred) # F1 Score : Best (1) Worst (0)
    precision = tpTrain/(tpTrain+fpTrain)
    recall = tpTrain/(tpTrain + fnTrain)

    # Calculate ROC
    dt_fpr, dt_tpr, dt_thresholds = roc_curve(actual, pred)
    score_roc_auc = auc(dt_fpr, dt_tpr)
    
    print(label)
    print()
    print("Accuracy :\t", accuracy)
    print()
    print("TPR Train  :\t", (tpTrain/(tpTrain + fnTrain)))
    print("TNR Train :\t", (tnTrain/(tnTrain + fpTrain)))
    print()
    print("FPR Train :\t", (fpTrain/(tnTrain + fpTrain)))
    print("FNR Train :\t", (fnTrain/(tpTrain + fnTrain)))
    print()

    print("Precision: \t", precision)
    print("Recall: \t", recall)
    print("F1-Score: \t", f1)
    print()
    print("{:.2f}".format(precision * 100) + "% of positive predictions made by the model were correct")
    print("{:.2f}".format(recall * 100) + "% of delays were correctly identified by the model")
    print()
    print("Area Under Curve (AUC) ROC Score: ", score_roc_auc)
    print()

    # Plot the Confusion Matrix
    sb.heatmap(confusion_matrix(actual, pred),
            annot = True, fmt=".0f", annot_kws={"size": 12})

    # Plot the ROC Curve
    plt.figure()
    plt.plot(dt_fpr, dt_tpr, color='darkorange', lw=2, label='ROC curve (AUC = %0.2f)' % score_roc_auc)
    plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('Receiver Operating Characteristic (ROC) Curve')
    plt.legend(loc="lower right")
    plt.show()
    model_scores[graph_label] = {"Accuracy":accuracy, "Precision":precision, "Recall":recall, "F1":f1, "AOC":score_roc_auc}
```

## 6. Discussion: Comparing Both Datasets <a class="anchor" id="sixth-bullet"></a>

```python
def showScoreDistribution(model_scores, title):
    
    # Extract model names
    model_names = list(model_scores.keys())
    
    # Extract score keys from the first model's scores dictionary
    score_keys = list(model_scores[model_names[0]].keys())

    # Set the width of the bars
    bar_width = 0.8 / len(score_keys)  # Adjust the width based on the number of scores
    
    # Set the positions of the bars on the x-axis
    bar_positions = [i + bar_width * j for i in range(len(model_names)) for j in range(len(score_keys))]
    
    # Create bar graph for each score
    plt.figure(figsize=(8, 4))
    for i, score_key in enumerate(score_keys):
        scores = [model_scores[model_name][score_key] for model_name in model_names]
        plt.bar(bar_positions[i::len(score_keys)], scores, width=bar_width, label=score_key)
    
    # Add x-axis labels and title
    plt.xlabel('Models', fontweight='bold')
    plt.xticks([r + (len(score_keys) - 1) * bar_width / 2 for r in range(len(model_names))], model_names)
    plt.title(title)
    
    # Add y-axis label
    plt.ylabel('Scores', fontweight='bold')
    
    # Add legend with appropriate keys
    plt.legend()
    
    # Show plot
    # Center the plot within the cell
    plt.tight_layout(pad=3.0)
    plt.show()


combined_scores = {**model_scores, **model_resampled_scores}
# Define the model names and scores you want to include in the subset
target_model_names = ["RF (Train)", "RF (Test)", "RF-R (Train)", "RF-R (Test)"]
target_scores = ["Accuracy", "Recall"]

# Create a new dictionary to store the subset
subset_model_scores = {
    model_name: {
        key: value for key, value in scores.items() 
        if key in target_scores
    } 
    for model_name, scores in combined_scores.items() 
    if model_name in target_model_names
}
```

## 7. Improving Our Model: Removing Irrelevant Features <a class="anchor" id="seventh-bullet"></a>

To reduce overfitting, we should identify **irrelevant features** that we are feeding our models with and remove them.
Thankfully, the decision tree models have a `feautre_importance` utility where features are scored based on their usefulness. 

Since we have previously one-hot encoded our categorical columns, we will **take the average of all columns with the same prefix** to give **each feature an overall importance score**.

```python
features = ['OperatingAirline', 'DayOfWeek', 'DayOfMonth', 'Dest', 'Origin', 'Hour', 'DistanceGroup']
# Get feature importances
feature_importances = model_rf_r.feature_importances_

# Initialize an empty dictionary to store average importance scores for each prefix
average_importance_scores = {}

# Iterate over each prefix
for prefix in features:
    # Filter columns with the current prefix
    prefix_columns = [col for col in X_train.columns if col.startswith(prefix)]
    
    # Calculate the average importance score of the filtered columns
    if prefix_columns:
        # Get the indices of columns with the current prefix
        prefix_indices = [list(X_train.columns).index(col) for col in prefix_columns]
        
        # Get the importance scores of columns with the current prefix
        prefix_importances = model_rf_r.feature_importances_[prefix_indices]
        
        # Calculate the average importance score of columns with the current prefix
        average_importance_score = np.mean(prefix_importances)
        average_importance_scores[prefix] = average_importance_score


# Sort the average importance scores dictionary by values in descending order
sorted_scores = sorted(average_importance_scores.items(), key=lambda x: x[1], reverse=True)

# Print the sorted average importance scores for each prefix
for prefix, score in sorted_scores:
    print("Average importance score of '{}': {:.6f}".format(prefix, score))
    
```

### Let us cross-check this list by plotting the top 30 most relevant features to our models.

```python
# Get feature importances
feature_importances = model_rf_r.feature_importances_

# Rank features based on importance scores
sorted_indices = np.argsort(feature_importances)[::-1]

# Select the top N most relevant features
top_n = 30  # Example: Select top 10 most relevant features
selected_features = X_train.columns[sorted_indices[:top_n]]
selected_importances = feature_importances[sorted_indices[:top_n]]

# Plot the bar graph
plt.figure(figsize=(10, 6))
plt.barh(selected_features, selected_importances, color='skyblue')
plt.xlabel('Importance Score')
plt.ylabel('Feature')
plt.title('Top 30 Most Relevant Features')
plt.gca().invert_yaxis()  # Invert y-axis to have the most important feature at the top
plt.show()
```

### **Keep `OperatingAirline`, Remove `Dest` and `Origin`**

Although `OperatingAirline` appears to be have a low average importance score, it is **one of the more relevant features** as shown in our graph. This could be because there are lot of flights from this particular airline. As such, we will keep this feature in and remove `Dest` and `Origin` since they have the **lowest average importance score** and **do not appear anywhere in the top 30 relevant features**.

```python
# Initialize an empty list to store selected columns
selected_columns = []

prefixes = ['OperatingAirline', 'DayOfWeek', 'DayOfMonth', 'Hour', 'DistanceGroup']

# Iterate over each prefix
for prefix in prefixes:
    # Filter columns that start with the current prefix
    prefix_columns = [col for col in df.columns if col.startswith(prefix)]
    
    # Extend the list of selected columns with columns that start with the current prefix
    selected_columns.extend(prefix_columns)

# Create a new DataFrame with only the selected columns
df_optimised = df[selected_columns]

# Split our data into train and test splits by 80-20 ratio, using random sampling.
# feature_set = pd.DataFrame(df_optimised.drop(labels=[label], axis=1))
# label_set = pd.DataFrame(df[label])
X_train, X_test, y_train, y_test = train_test_split(df_optimised, label_set, test_size=0.2, random_state=randomStateNumber)
```

## 8. Improving Our Model: Tuning Hyperparameters <a class="anchor" id="eigth-bullet"></a>

To further optimise the performance of the models that we have created, we can tune the hyperparameters we used to create it. <br>This is akin to **tweaking our computer's configuration to make it run faster**.

How can we identify what hyperparameters to use? There are two popular ways: `GridSearch` and `RandomizedSearch`.

> `GridSearch`: Grid search is a simple and straightforward method that involves **defining a grid of hyperparameter values and evaluating each combination using cross-validation**. Grid search exhaustively searches through all possible combinations of hyperparameters and identifies the combination that yields the best performance.   

> `RandomizedSearch`: Randomized search is an alternative to grid search that **samples hyperparameter values from specified probability distributions**. Unlike grid search, which evaluates all possible combinations, randomized search evaluates a fixed number of random combinations. This approach can be more efficient when the hyperparameter search space is large.



For simplicity sake, we will use `GridSearch` for this project.

### Supplementing with k-Fold Cross Validation

To ensure that the results of our model are not purely conincidental with the splits we have randomly sampled, we will also employ `k-Fold Cross Validation` to find a `cross validation score`.

By **splitting the dataset into `k groups`**, we repeatedly select the `i-th` sample as a test group and the remaining `k-1 groups` to be training sets to train on. Finally, we take the **average evaluation score for all `k groups`** as the overall score for the model.

With this, we will be able to **more accurately represent the effectiveness of the model used**, knowing that there is **minimal sampling bias** from generating the train/test splits.

### Combining `GridSearch` and `Cross Validation`

Let us combine both `GridSearch` and `Cross Validation` to find the best hyperparameters for the models that we have trained so far and score them to find the model with the best `recall` score.
We chose to use `recall` here instead of `accuracy` as we want to be able to **adequately prepare passengers for potential delays (reduce false negatives)**, instead of prioritising unnecessary decisions in which case precision would be preferred.

### **DecisionTreeClassifier (Tuned)**

```python
model_tuned = {}
```

```python
# Define the parameter grid
param_grid_dt = {
    'max_depth': [None, 5, 10, 15],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

grid_search_dt = GridSearchCV(model_dt_r, param_grid_dt, cv=5, scoring='recall')
grid_search_dt.fit(X_train, y_train)

# Get the best parameters and best score
print("Best Parameters:", grid_search_dt.best_params_)
print("Best Recall Score:", grid_search_dt.best_score_)

# Evaluate the best model on the test set
best_dt = grid_search_dt.best_estimator_
best_dt_y_pred = best_dt.predict(X_test)

printMetrics(model_tuned, "DT (Tuned)", "DecisionTreeClassifier (Tuned)", y_test, best_dt_y_pred)
```

### **RandomForestClassifier (Tuned)**

Very interestingly, the features most relevant to the model were flights from **Southwest Airlines** (`OperatingAirline_WN`),  **Skywest Airlines**  (`OperatingAirline_OO`) and flights from **5AM to 8AM**. Meaning, most flights that were from **Southwest Airlines or from 5-8AM** had a higher chance of delay.

A quick check back on the news in 2022 shows that there were **many disruptions from Southwest Airlines** at that point in time. 

Despite our efforts, the performance of our model is far from good enough in accurately predicting the presence of a delayed flight in reality.

This could be due to several factors:

### **Feature Relevance**
The features used for model training are not adequate enough in capturing factors contributing to flight delays. Although we have implemented feature selection techniques to identify the most relevant predictors, there could still be overlooked variables that are important in accurate predictions but are not represented adequately in the dataset for it to be chosen.

### **Nature of Flight Delay and Limitation of Dataset**
Flight delays in reality are often influenced by external events like **severe weather and service disruptions**. However, these information are not available in the dataset chosen. Thus, **augmenting the dataset with real-time forecasts** of such factors can significantly improve model performance.
